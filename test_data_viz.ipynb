```python
import pandas as pd 
import numpy as np 
import pickle
bst = pickle.load(open('lgbm_2.sav', 'rb'))
```

```python
test_data=pd.read_csv('test_data_for_viz.csv')
final_features = ['Situational xPass','FAD',"# of RB's",'RMV','TICK','Run Formation Flag','DCI']
test_data['RPO_pickle']=bst.predict_proba(test_data[final_features])[:, 1]
```

```python


test_agg=test_data[['possessionTeam', 'playId', 'pass']].groupby('possessionTeam').agg({'playId':"count", 'pass':"sum"} ).reset_index()
test_agg['pass_rate']=test_agg['pass']/test_agg['playId']

test_pred=test_data[['possessionTeam', 'Situational xPass', 'RPO', 'RPO_pickle']].groupby('possessionTeam').mean().reset_index()
test=pd.merge(test_agg, test_pred, on='possessionTeam')
```

```python
np.corrcoef(test['pass_rate'], test['Situational xPass'])
```

```python
np.corrcoef(test['pass_rate'], test['RPO'])
```

```python
test['delta_xpass']=test['pass_rate']-test['Situational xPass']
test['delta_RPO']=test['pass_rate']-test['RPO']
```

```python
test[['delta_xpass', 'delta_RPO']].mean()
```
